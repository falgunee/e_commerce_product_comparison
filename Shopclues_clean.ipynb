```python
import pandas as pd
fd=pd.read_excel("Shopcluesfile.xlsx")
fd.head(5)
```

```python
fd['Price'] = (fd['Price'].str[3:].astype(str))
fd.head(4)
```

```python
s=fd['Names']
type(s)
a=pd.DataFrame(s.apply(lambda st: st[st.find("(")+1:st.find(")")]))
a.head(5)
```

```python
a['RAM'] = a.Names.str.split(',').str[0].astype(str)
a['STORAGE'] = a.Names.str.split(',').str[1].astype(str)
a['Colour'] = a.Names.str.split(',').str[2].astype(str)
a.head(5)
```

```python
fd["Names"]= fd["Names"].str.split("(",expand=True)
fd.head(4)
fd['Brand_Name'] = fd.Names.str.split(' ', 1).str[0].astype(str)
fd['Model_Name']=fd.Names.str.split(' ', 1).str[1].astype(str)
fd.head(4)
```

```python
main_df=pd.concat([fd["Names"],fd["Brand_Name"],fd["Model_Name"],a["RAM"],a["STORAGE"],fd["Price"],fd["Discount"],fd["Links"],a['Colour']],axis=1)
main_df.head(20)
```

```python
writer = pd.ExcelWriter(r'Main_clean_shopclues_finaL.xlsx', engine='xlsxwriter',options={'strings_to_urls': False})
main_df.to_excel(writer)
writer.close()
```
