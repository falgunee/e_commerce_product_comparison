```python
import pandas as pd
fd=pd.read_excel("Reliancefile.xlsx")
fd.head(20)
```

```python
fd['Price']=fd.Price.str.split('₹', 1).str[1].astype(str)
fd["Price"]
```

```python
s=fd['Names']
a=pd.DataFrame(s.apply(lambda st: st[st.find("[A-za-z]")+1:st.find(",")]))
a.head(4)
```

```python
a['STORAGE'] = a['Names'].str.extract('(\d?\d\d\sGB)', expand=True)
#a['STORAGE']
```

```python
fd["RAM"]=fd["Names"].str.extract('(\d\sGB\sRAM)', expand=True)
#fd["RAM"]
```

```python
fd['Brand_Name'] = fd.Names.str.split(' ', 1).str[0].astype(str)
fd.head(5)
```

```python
fd['Model_Name']=fd.Names.str.split(' ', 1).str[1].astype(str)
fd['Model_Name'] = fd.Model_Name.str.split('\s\d\sGB').str[0].astype(str)
fd.head(4)
```

```python
a=pd.DataFrame(fd["Model_Name"].apply(lambda st: st[st.find(""):st.find("Smart")]))
a.head(20)
fd["Model_Name"]=a["Model_Name"]
fd.head(4)
```

```python
main_df=pd.concat([fd["Model_Name"],fd['Brand_Name'],fd["RAM"],a["STORAGE"],fd["Links"],fd["Price"]],axis=1)
main_df.head(4)
```

```python
fd["Colour"]=fd["Names"].str.split(',\s\d\sGB\sRAM,\s').str[1].astype(str)
fd["Colour"]
```

```python
df1=pd.DataFrame(fd["Brand_Name"]+" "+fd["Model_Name"],columns=["Names"])
df1
final_df=pd.concat([df1,fd['Brand_Name'],fd["Model_Name"],fd["Price"],fd["RAM"],a["STORAGE"],fd["Colour"],fd["Links"]],axis=1)
final_df
```

```python
#we use this method to write excel file because length of url was too long to fit
writer = pd.ExcelWriter(r'Relianceclean.xlsx', engine='xlsxwriter',options={'strings_to_urls': False})
final_df.to_excel(writer)
writer.close()
```
