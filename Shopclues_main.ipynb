```python
from selenium import webdriver
import time
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
driver=webdriver.Chrome()
url="https://www.shopclues.com/search?q=smartphones&sc_z=2222&z=1&show_brand[]=1455062&show_brand[]=826485&show_brand[]=1455038&show_brand[]=923724&show_brand[]=1519075&by_item_condition[]=2153751&hcat_id=1431&show_brand[]=1519082&show_brand[]=1278511&show_brand[]=1455475&show_brand[]=1455623&show_brand[]=1454974&show_brand[]=105751&fsrc=show_brand,by_item_condition"
driver.get(url)
time.sleep(30)
driver.execute_script("window.scrollBy(0,2000)")
wait = WebDriverWait(driver, 10)
while True:
    try:
        loadmore = wait.until(EC.visibility_of_element_located((By.CSS_SELECTOR,"[id='moreProduct']")))
        driver.execute_script("arguments[0].click();", loadmore)
        wait.until(EC.staleness_of(loadmore))
    except Exception:break
```

```python
names=[]
page=driver.find_elements_by_tag_name('h2')
for i in range(len(page)):
    a=page[i].text
    names.append(a)
    print(names)
    
```

```python
# getting the price 
prices=[]
price=driver.find_elements_by_class_name('p_price')
for i in range(len(price)):
    a=price[i].text
    prices.append(a)
    
```

```python
# getting the discount
discounts=[]
discount=driver.find_elements_by_class_name('prd_discount')
for i in range(len(discount)):
    a=discount[i].text
    discounts.append(a)
    #print(discounts)
    
```

```python
import urllib
from bs4 import BeautifulSoup
import requests
source = driver.page_source
soup = BeautifulSoup(source,'html.parser') 
li = []
link=soup.findAll("div",attrs={'class':'column col3 search_blocks'})
for elem in link:
    a = elem.findAll("a")
    li.append(a[0].get('href'))
    print(li)
```

```python
import pandas as pd
n=pd.DataFrame(names)
p=pd.DataFrame(prices)
d=pd.DataFrame(discounts)
l=pd.DataFrame(li)
```

```python
main_df=pd.concat([n,p,d,l],axis=1)
my_col=["Names","Price","Discount","Links"]
main_df.columns=my_col
main_df
```

```python
writer = pd.ExcelWriter(r'Shopcluesfile.xlsx', engine='xlsxwriter',options={'strings_to_urls': False})
main_df.to_excel(writer)
writer.close()
```
