```python
from selenium import webdriver
import time
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
driver = webdriver.Chrome()
url = 'https://www.snapdeal.com/products/mobiles-mobile-phones/filters/Form_s~Smartphones?sort=plrty&q=Form_s%3ASmartphones%7CBrand%3ARedmi%5EApple%5EHonor%5ELenovo%5ELG%5EMI%5EMoto%5EOnePlus%5EOppo%5ERedmi%20Note%205%20Pro%5ESamsung%20Mobile%5EVivo%7C'
driver.get(url)
time.sleep(30)
driver.execute_script("window.scrollBy(0,2000)")
time.sleep(30)
```

```python
source = driver.page_source
soup = BeautifulSoup(source,'html.parser')
link = soup.findAll('div',attrs={'class':'product-tuple-image'})
li = []
for elem in link:
    a = elem.findAll("a",attrs={'class':'dp-widget-link'})
    #print(a[0].get('href'))
    li.append(a[0].get('href'))
    print(li)
```

```python
# getting device names
names=[]
page = driver.find_elements_by_class_name('product-title')
for i in range(len(page)):
    a=page[i].text
    names.append(a)
    print(names)
```

```python
# getting the price 
prices=[]
price=driver.find_elements_by_class_name('product-price')
for i in range(len(price)):
    a=price[i].text
    prices.append(a)  
df_prices=pd.DataFrame(prices)
df_prices.shape    
    
```

```python
# getting the discount
discounts=[]
discount=driver.find_elements_by_class_name('product-discount')
for i in range(len(discount)):
    a=discount[i].text
    discounts.append(a)
df_discounts=pd.DataFrame(discounts)
df_discounts.shape     
```

```python
n=pd.DataFrame(names)
p=pd.DataFrame(prices)
d=pd.DataFrame(discounts)
l=pd.DataFrame(li)
```

```python
main_df=pd.concat([n,p,d,l],axis=1)
```

```python
my_col=["Names","Price","Discount","Links"]
main_df.columns=my_col
main_df
```

```python
writer = pd.ExcelWriter(r'Snapdealfile.xlsx', engine='xlsxwriter',options={'strings_to_urls': False})
main_df.to_excel(writer)
writer.close()
```
