```python
from selenium import webdriver
import requests
from bs4 import BeautifulSoup
driver=webdriver.Chrome()
url='https://www.tatacliq.com/electronics-tablets-tablets/c-msh1210100?q=%3Arelevance%3Acategory%3AMSH1210100%3Acategory%3AMSH1210100%3Acategory%3AMSH1210100%3Acategory%3AMSH1210100%3Acategory%3AMSH1210100%3Acategory%3AMSH1210100%3Acategory%3AMSH1210100%3Acategory%3AMSH1210100%3Acategory%3AMSH1210100%3Acategory%3AMSH1210100%3AinStockFlag%3Atrue%3AisLuxuryProduct%3Afalse%3Abrand%3AMBH12E01025%3Abrand%3AMBH12E00007%3Abrand%3AMBH12E00018%3Abrand%3AMBH12E00076%3Abrand%3AMBH12E00130%3Abrand%3AMBH12E00061%3Abrand%3AMBH12E00256%3Abrand%3AMBH12E00016%3Abrand%3AMBH12E00181%3Abrand%3AMBH12E00015'
driver.get(url)
#manually loaded all the prodcuts
```

```python
names=[]
prices=[]
links=[]
offers=[]
page = driver.find_elements_by_class_name("Bt9jWqBhJEHlqtj4x2xNa")
for i in range(len(page)):
    a=page[i].text
    names.append(a)

    
price=driver.find_elements_by_class_name('dKCupF_5rtgdFHZiQ3xpQ') 
for i in range(len(price)):
    a=price[i].text
    prices.append(a)

    
pgsource = str(driver.page_source)
soup = BeautifulSoup(pgsource, 'html5lib')
tag = soup.findAll('div',{"class":"_26uHe2QbvnuoUvswPtzSnX"})
for t in tag:
    if t.find('a') != None:
        link = t.find('a').get('href')
        links.append(link)
        
        
offer=driver.find_elements_by_class_name("_1IxIox0fPQHTc_jS_sokZD")
for i in range(len(offer)):
    a=offer[i].text
    offers.append(a)
    
```

```python
import pandas as pd

df_names=pd.DataFrame(names)
df_links=pd.DataFrame(links)
df_prices=pd.DataFrame(prices)
df_offer=pd.DataFrame(offers)


print(df_names.shape)
print(df_links.shape)
print(df_prices.shape)
print(df_offer.shape)
```

```python
frames=[df_names,df_prices,df_links,df_offer]
mycol=["Names","Prices","Links","Offer"]
main_df1=pd.concat(frames,axis=1)
main_df1.columns=mycol
main_df1.dropna(axis=0)
main_df1.head(25)
```

```python
writer = pd.ExcelWriter(r'tatacliq_file.xlsx', engine='xlsxwriter',options={'strings_to_urls': False})
main_df1.to_excel(writer)
writer.close()
```
