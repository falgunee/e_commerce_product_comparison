```python
import pandas as pd
import numpy as npip
import seaborn as sns
import matplotlib
from matplotlib import pyplot as plt
from bokeh.io import output_notebook
from bokeh.plotting import figure, show
```

```python
df=pd.read_excel("Main_clean_Amazon_final.xlsx")
a=df.loc[df['RAM'].isin([4,3,16,8,6,2,32,64])]
#df["STORAGE"].value_counts()
b=df.loc[df['STORAGE'].isin([64,32,128,16,8])]
```

```python
plt.figure(figsize=(15,5))
data=pd.DataFrame(df.groupby('Brand_Name')['Prices'].max().sort_values(ascending=False)).reset_index()
rep_plot=sns.barplot(x='Brand_Name',y='Prices',data=data,palette=sns.set_palette(palette='viridis_r'))
rep_plot.set_xlabel("BrandName")
rep_plot.set_ylabel("Max_Price ")
rep_plot.set_title('"Max_Price for top model')
```

```python
plt.figure(figsize=(15,5))
data=pd.DataFrame(df.groupby('Brand_Name')['Prices'].mean().sort_values(ascending=False)).reset_index()
ax=sns.barplot(x='Brand_Name',y='Prices',data=data,palette=sns.set_palette(palette='viridis_r'))
ax.set_xlabel('Brand_Name')
ax.set_title("Average price of every Brand")
```

```python
plt.figure(figsize=(15,5))
data=pd.DataFrame(a.groupby('Brand_Name')['RAM'].max().sort_values(ascending=False)).reset_index()
rep_plot=sns.barplot(x='Brand_Name',y='RAM',data=data,palette=sns.set_palette(palette='viridis_r'))
rep_plot.set_xlabel("BrandName")
rep_plot.set_ylabel("Max_RAM")
rep_plot.set_title('Max RAM - Brandwise')
```

```python
plt.figure(figsize=(15,5))
data=pd.DataFrame(b.groupby('Brand_Name')['STORAGE'].max().sort_values(ascending=False)).reset_index()
rep_plot=sns.barplot(x='Brand_Name',y='STORAGE',data=data,palette=sns.set_palette(palette='viridis_r'))
rep_plot.set_xlabel("BrandName")
rep_plot.set_ylabel("Max_STORAGE")
rep_plot.set_title('Max storage capacity - Brandwise')
rep_plot.set_ylim((0,128))
```

```python
df['Brand_Name'].value_counts().plot.bar(title="Max Model Types")
```

```python
plt.figure(figsize=(15,5))
data=pd.DataFrame(df.groupby('Brand_Name')['Prices'].sum().sort_values(ascending=False)).reset_index()
ax=sns.barplot(x='Brand_Name',y='Prices',data=data,palette=sns.set_palette(palette='viridis_r'))
#ax.ticklabel_format()
ax.set_xlabel('Brand_Name')
ax.set_ylabel('Totalprice for all the models')
ax.set_title('List of pricey mobiles for all models')
```
