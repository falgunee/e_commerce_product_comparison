```python
import re
names=[]
links=[]
all_info=[]
prod_lst=[]
price_lst=[]
for u in url:
    driver.get(u)
    posts=driver.find_elements_by_class_name("s-item-container")


    cnt=1
    for post in posts:
        ptext=post.text
        pind=ptext.find('by')
        pmatch=(ptext[:pind-1])
        prod_lst.append(pmatch)
        prmatch=re.search('(\d{1,},\d{1,})',ptext[pind:])
        if prmatch!=None:
            price=prmatch.group(1)
        else:
            prmatch=re.search('(\d{1,}) ',ptext[pind:])
            if prmatch!=None:
                price=prmatch.group(1)
        price_lst.append(price)
        cnt=cnt+1
        if cnt==len(posts):
            break
    #print(prod_lst)
    #print(price_lst)
   
        
    pgsource = str(driver.page_source)
    soup = BeautifulSoup(pgsource, 'html5lib')
    tag = soup.findAll('div',{"class":"s-item-container"})
    for t in tag:
        if t.find('a') != None:
            link = t.find('a').get('href')
            links.append(link)
    #print(links)
    
    info=driver.find_elements_by_class_name("a-definition-list ")
    for i in range(len(info)):
        a=info[i].text
        all_info.append(a)
    #print(all_info)

    
    
```
